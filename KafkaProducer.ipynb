```python
import pandas as pd
from kafka import KafkaProducer
from time import sleep
from json import dumps
import json
import os
from dotenv import load_dotenv
import boto3
```

```python
load_dotenv()

producer = KafkaProducer(bootstrap_servers=os.getenv("KAFKA_BOOTSTRAP_SERVERS"),
                         value_serializer=lambda x: dumps(x).encode('utf-8'))
```

```python
producer.send(os.getenv('KAFKA_TOPIC'), value="{'hello':'world'}")
```

```python
chunk_size=100
for chunk in pd.read_csv('indexProcessed.csv', chunksize=chunk_size):
    records = chunk.to_dict(orient='records')

    for record in records:
        producer.send(os.getenv('KAFKA_TOPIC'), value=record)

    print(f"Sent {len(records)} rows from csv file (chunk)")
```

```python
producer.flush()
```
